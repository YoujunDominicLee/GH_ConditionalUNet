```python
import time
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split, Subset, ConcatDataset
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
import random
import pickle


SEED=0
torch.manual_seed(SEED)
random.seed(SEED)
np.random.seed(SEED)
```

```python
class NumpyDataset(Dataset):
    def __init__(self, x_path, y_path):
        self.x = np.load(x_path)
        self.y = np.load(y_path)
    
    def __len__(self):
        return self.x.shape[0]
    
    def __getitem__(self, idx):
        x_full = self.x[idx]
        main_x = torch.tensor(x_full[:12, :, :], dtype=torch.float32)
        time_enc = torch.tensor(x_full[12:13, :, :], dtype=torch.float32)
        y = torch.tensor(self.y[idx], dtype=torch.float32)
        return main_x, time_enc, y
```

```python
for i in range(1,8):
    train1= torch.load(f'training_data/train_dataset_{i}.pt')
    val1  = torch.load(f'training_data/val_dataset_{i}.pt')
    test1 = torch.load(f'training_data/test_dataset_{i}.pt')
    
    for j in range((i+1),9):
        train2= torch.load(f'training_data/train_dataset_{j}.pt')
        val2  = torch.load(f'training_data/val_dataset_{j}.pt')
        test2 = torch.load(f'training_data/test_dataset_{j}.pt')

        train_dataset = ConcatDataset([train1, train2])
        val_dataset = ConcatDataset([val1, val2])
        test_dataset = ConcatDataset([test1, test2])
        
        torch.save(train_dataset, f'training_data/train_dataset_{i}_{j}.pt', pickle_protocol=pickle.HIGHEST_PROTOCOL)
        torch.save(val_dataset, f'training_data/val_dataset_{i}_{j}.pt', pickle_protocol=pickle.HIGHEST_PROTOCOL)
        torch.save(test_dataset, f'training_data/test_dataset_{i}_{j}.pt', pickle_protocol=pickle.HIGHEST_PROTOCOL)
```
